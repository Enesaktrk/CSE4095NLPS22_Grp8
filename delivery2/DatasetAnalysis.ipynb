```python
import matplotlib.pyplot as plt
import json
```

```python
# Load dataset
with open('dataset.json') as json_file:
    dataset = json.load(json_file)
```

```python
occurrences = dict()
for data in dataset:
    try:
        occurrences[dataset[data][0]] += 1
    except:
        occurrences[dataset[data][0]] = 1

sorted_values = sorted(occurrences.values(), reverse=True) # Sort the values
sorted_dict = {}

for i in sorted_values:
    for k in occurrences.keys():
        if occurrences[k] == i:
            sorted_dict[k] = occurrences[k]
            break
```

```python
for label in sorted_dict:
    print(f'{label}: {sorted_dict[label]}')
```

```python
# Plot frequencies
plt.bar(sorted_dict.keys(), sorted_dict.values())
plt.xticks(rotation='vertical')
plt.show()
```
